```python
import numpy as np 
import scipy
from scipy.integrate import quad
from matplotlib import pyplot as plt

def integrand(x, d):
    return x * scipy.special.jn(0, d*x)

xline = np.linspace(0, 1, 1000)

vals = []
for x in xline:
    vals.append(quad(integrand, 0, 7, args=(x))[0])
plt.plot(xline, vals, linewidth=2.0, label=r'$\Lambda_1$', color='mediumslateblue')

vals = []
for x in xline:
    vals.append(quad(integrand, 0, 14, args=(x))[0])
plt.plot(xline, vals, label=r'$\Lambda_2 = 2\Lambda_1$', linewidth=2.0, color='deepskyblue')

vals = []
for x in xline:
    vals.append(quad(integrand, 0, 28, args=(x))[0])
plt.plot(xline, vals, label=r'$\Lambda_3 = 4\Lambda_1$', linewidth=2.0, color='mediumaquamarine')

plt.plot([0,1], [0,0], '--', color='black', linewidth=2.0)

'''plt.xticks([1/7, 1/12, 1/20])
ax = plt.gca()
ax.set_xticklabels(["a1", "a2", "a3"])'''
plt.xticks([])
plt.yticks([0], fontsize=12)
plt.legend(fontsize=14)
plt.xlabel("d =|x-y|", fontsize=14)
plt.ylabel(r"$\left\langle\eta(x) \, \eta(y)\right\rangle$", fontsize=14)

plt.tight_layout()
plt.savefig("bessel.pdf")
plt.show()
```
