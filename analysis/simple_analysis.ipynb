```python
import numpy as np
from matplotlib import pyplot as plt 
from pandas import read_csv
import sys
import os
from matplotlib import rcParams
import toml

from correlations import *
from read_in_data import *
from mass_fit import *
from mass_matteo import *
from functions import *

%load_ext autoreload
%autoreload 2
```

```python
param = ("physics", "mass")
simulations = []
for i in [6]:
    dataset = Dataset(128, 128)
    dataset.add_data("./data/looking4pt//conf" + str(i) + "/", param, 1)
    simulations.append(dataset)
```

```python
print("abs. phi:")
data = []
for s in simulations:
    val, err = s.compute_mag(printing=False)
    print(s.toml_params[param[0]][param[1]], val, err)
    data.append((s.toml_params[param[0]][param[1]], val, err))
params, vals, errs = sort_data(data)
plt.plot(params, vals, 'o--')
plt.xlabel(r"$m_\phi^2$")
plt.ylabel(r"$\left\langle|\phi|\right\rangle$")
plt.tight_layout()
plt.savefig("magnetisation.pdf")
plt.grid()
plt.show()
```

```python
print("cond. :")
data = []
for s in simulations:
    cF = s.toml_params['fermions']['fermion_mass']
    val, err = s.compute_condensate(printing=False)
    print(s.toml_params[param[0]][param[1]], cF*val, cF*err)
    data.append((s.toml_params[param[0]][param[1]], val, err))
params, vals, errs = sort_data(data)
plt.plot(params, vals, 'o--')
plt.xlabel(r"$m_\phi^2$")
plt.ylabel(r"$\left\langle\bar\psi\psi\right\rangle$")
plt.savefig("condensate.pdf")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
print("chi2:")
data = []
for s in simulations:
    val, err = s.compute_susceptibility(printing=False)
    print(s.toml_params[param[0]][param[1]], val, err)
    data.append((s.toml_params[param[0]][param[1]], val, err))
    data.append((s.toml_params[param[0]][param[1]], val, err))
params, vals, errs = sort_data(data)
plt.plot(params, vals, 'o--')
plt.xlabel(r"$m_\phi^2$")
plt.ylabel(r"$\left\langle\chi^2\right\rangle$")
plt.tight_layout()
plt.grid()
plt.savefig("susceptibility.pdf")
plt.show()
```

```python
print("mass:")
data = []
for s in simulations:
    try:
        val, err = s.compute_mqphys(plotting=True)
        print(s.toml_params[param[0]][param[1]], val, err)
        data.append((s.toml_params[param[0]][param[1]], val, err))
        data.append((s.toml_params[param[0]][param[1]], val, err))
    except:
        pass
params, vals, errs = sort_data(data)
plt.plot(params, vals, 'o--', markersize=3)
plt.xlabel(r"$m_\phi^2$")
plt.ylabel(r"$m_q^{phys}$")
plt.tight_layout()
plt.grid()
plt.savefig("mass.pdf")
plt.show()
```
